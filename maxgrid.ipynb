```python
import math

def findMaxIndex(mat, col):
    """Finds the row index of the maximum element in a given column."""
    maxRow = 0
    n = len(mat)
    for i in range(1, n):
        if mat[i][col] > mat[maxRow][col]:
            maxRow = i
    return maxRow

def findPeakGrid(mat):
    """Finds a peak element in a 2D matrix using binary search on columns."""
    n = len(mat)
    m = len(mat[0])
    low, high = 0, m - 1

    while low <= high:
        midCol = (low + high) // 2
        maxRow = findMaxIndex(mat, midCol)
        currentElement = mat[maxRow][midCol]

        # Get left and right neighbors safely (treating out-of-bounds as -1 or negative infinity)
        leftNeighbor = mat[maxRow][midCol - 1] if midCol > 0 else -1
        rightNeighbor = mat[maxRow][midCol + 1] if midCol < m - 1 else -1

        # Check if the current element is a peak
        if currentElement > leftNeighbor and currentElement > rightNeighbor:
            return [maxRow, midCol]
        # If left neighbor is bigger, move search to the left half
        elif currentElement < leftNeighbor:
            high = midCol - 1
        # If right neighbor is bigger, move search to the right half
        else:
            low = midCol + 1

    return [-1, -1] # Should not reach here if a peak is guaranteed to exist

# Example Usage:
mat = [
    [10, 8, 10],
    [14, 13, 12],
    [15, 9, 11]
]

peakLocation = findPeakGrid(mat)
print(f"Peak found at coordinates (row, col): {peakLocation}")
```
