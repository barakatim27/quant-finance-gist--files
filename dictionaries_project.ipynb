```python
# Student Score Tracker
# 1.Store student name and student score.
# 2.Print all students.
# 3.Calculate average score.

my_dict = {
    1: {
        "name" : "Timothy Baraka",
        "score" : 95
        },
    2: {
        "name" : "John Doe",
         "score" : 80
        },
     3: {
        "name" : "Jane Doe",
        "score" : 40},
     4: {
        "name" : "Tigana Ochieng",
        "score" : 50
        },
     5: {"name": "Troy Mwamba",
         "score": 40
        }
}
print(my_dict)

total_score = 0

for x in my_dict.values():
    total_score += x["score"]

print(total_score)

average_score = total_score / len(my_dict)

print(average_score)
```
