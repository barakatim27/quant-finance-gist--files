```python
thisdict = {
    'brand':'Ford',
    'model': 'Mustang',
    'year' : 1964
}
print(thisdict)
```

```python
thisdict = {
    'brand':'Ford',
    'model': 'Mustang',
    'year' : 1964,
    'year': 2020
}
print(thisdict)
```

```python
# len and type
thisdict = {
    'brand':'Ford',
    'model': 'Mustang',
    'year' : 1964,
}
print(len(thisdict))
print(type(thisdict))
```

```python
#using a constructor
thisdict = dict(name = 'john', age = 36, country = 'Norway')
print(thisdict)
```

```python
# accessing items
thisdict ={
    'brand':'Ford',
    'model': 'Mustang',
    'year' : 1964,
}
a = thisdict['model']
print(a)

#get method
b = thisdict.get("model")
print(b)

# find keys
c = thisdict.keys()
print(c)

#find values
d = thisdict.values()
print(d)

#get all items
e = thisdict.items()
print(e)
```

```python
thisdict  = {
    'brand':'Ford',
    'model': 'Mustang',
    'year' : 1964,
}
if "model" in thisdict:
    print("yes")
```

```python
#update dict
thisdict  = {
    'brand':'Ford',
    'model': 'Mustang',
    'year' : 1964,
}
thisdict.update({"year": 2020})
print(thisdict)
```

```python
thisdict  = {
    'brand':'Ford',
    'model': 'Mustang',
    'year': 1964
}
thisdict["color"] = "red"
print(thisdict)
```

```python
thisdict ={
    'brand':'Ford',
    'model': 'Mustang',
    'year' : 1964
}

thisdict.update({"color": "red"})
print(thisdict)
```

```python
# remove item
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict.pop("model")
print(thisdict)
```

```python
#remove the last item
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict.popitem()
print(thisdict)
```

```python
# delete key
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
del thisdict["model"]
print(thisdict)
```

```python
# delete the dict
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
del thisdict
print(thisdict)
```
