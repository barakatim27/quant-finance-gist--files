```python
numbers = [10, 15, 20, 25]

total = 0

for num in numbers:
   total += num

mean = total / len(numbers)

print("Numbers:", numbers)
print("Mean:", mean)
print("Max:",max(numbers))
print("Min:",min(numbers))
```
