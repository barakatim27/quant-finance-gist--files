```python
fruits = ['apple', 'mango', 'banana', 'cherry']
for x in fruits:
  print(x)
  if x == 'banana':
    break
```

```python
fruits = ['apple', 'mango', 'banana', 'cherry']
for x in fruits:
  if x == 'banana':
     break
  print(x)
```

```python
fruits = ['apple', 'mango', 'banana', 'cherry']
for x in fruits:
  if x == 'banana':
     continue
  print(x)
```
