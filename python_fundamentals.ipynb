```python
# Variables can store data of various types, and different types can do
# different things.

# str --> text type
my_str = "Hello"
print(my_str)
```

```python
# Numeric types
my_int  = 20
my_float  = 1.5
my_complex = 1j

print(my_int)
print(my_float)
print(my_complex)
```

```python
# Sequence types
my_list  = [1, 2, 3, 4, 5]
my_tuple = (1, 2, 3, 4, 5)
my_range  = range(100)

print(my_list)
print(my_tuple)
print(my_range)
```

```python
# Mapping type
my_dict ={'name' : 'timothy baraka', 'age': 27}
print(my_dict)
```

```python
# set types
my_set  = {1, 2, 3, 4, 5}
my_frozenset = frozenset({1, 2, 3, 4, 5})

print(my_set)
print(my_frozenset)
```

```python
# boolean type
my_bool = True

print(my_bool)
```

```python
# None
x  = None
print(x)
```
